```python
for width in [1,3,5,8]:
    for offset in [0,2,4]:
        mask = 2 ** (width + offset) - 2 ** offset 
        mask = mask & 255
        print(bin(mask))
```

```python
reg_size = 8
bits = [7, 3]
val = 16

msb = bits[0]
lsb = bits[1]
val = val * 2 ** lsb
val = val % (2 ** reg_size)
mask = 2 ** (msb + 1) - 2 ** lsb

print(bin(val), bin(mask))
```
